```python
import pandas as pd
from WikipediaAnalytics import WikipediaAnalytics
```

```python
def test_code(wa : WikipediaAnalytics):
    # ===============================================================
    # 1) TESTS: Selección de filas
    # ===============================================================

    print("📌 Testing select_row_by_value...")

    # Test 1: Coincidencia correcta
    row = wa.select_row_by_value("Country Name", "España")
    assert row is not None
    assert row.iloc[0]["Population (hab.)"] == 49570725.0
    print("  ✔ select_row_by_value encuentra una fila existente.")

    # Test 2: No coincidencia
    row = wa.select_row_by_value("Country Name", "Francia")
    assert row is None
    print("  ✔ select_row_by_value devuelve None cuando no hay coincidencia.")

    # Test 3: Columna inválida
    try:
        wa.select_row_by_value("NoExiste", "España")
        print("  ❌ ERROR: no lanzó excepción con columna inválida")
    except ValueError:
        print("  ✔ select_row_by_value detecta columnas inválidas.")


    # ===============================================================
    # 2) TESTS: Selección de columnas
    # ===============================================================

    print("\n📌 Testing get_columns...")

    # Test 1: Una sola columna
    col = wa.get_columns("GDP ($)")
    assert list(col.columns) == ["GDP ($)"]
    print("  ✔ get_columns devuelve una columna correctamente.")

    # Test 2: Varias columnas
    cols = wa.get_columns(["Country Name", "Population (hab.)"])
    assert list(cols.columns) == ["Country Name", "Population (hab.)"]
    print("  ✔ get_columns devuelve varias columnas correctamente.")

    # Test 3: Columna inexistente
    try:
        wa.get_columns(["GDP ($)", "NOPE"])
        print("  ❌ ERROR: no lanzó excepción con columna inválida")
    except ValueError:
        print("  ✔ get_columns detecta columnas inválidas.")


    # ===============================================================
    # 3) TESTS: Funciones de agregación
    # ===============================================================

    print("\n📌 Testing aggregate_column...")

    # Test max
    max_pop = wa.aggregate_column("Population (hab.)", "max")
    assert max_pop == 49570725.0
    print("  ✔ aggregate_column calcula correctamente el máximo.")

    # Test min
    min_density = wa.aggregate_column("Density (hab./km^2)", "min")
    assert min_density == 97.47
    print("  ✔ aggregate_column calcula correctamente el mínimo.")

    # Test mean
    mean_lat = wa.aggregate_column("Latitude (º)", "mean")
    expected = (44.818 + 52.230 + 40.417) / 3
    assert abs(mean_lat - expected) < 1e-6
    print("  ✔ aggregate_column calcula correctamente la media.")

    # Test operación inválida
    try:
        wa.aggregate_column("GDP ($)", "median")
        print("  ❌ ERROR: no lanzó excepción con operación inválida")
    except ValueError:
        print("  ✔ aggregate_column detecta operaciones inválidas.")

    # Test columna inexistente
    try:
        wa.aggregate_column("NOPE", "max")
        print("  ❌ ERROR: no lanzó excepción con columna inexistente")
    except ValueError:
        print("  ✔ aggregate_column detecta columnas inexistentes.")


    print("\n🎉 TODOS LOS TESTS HAN SIDO EJECUTADOS CORRECTAMENTE.")
```

```python
# Prueba Base
df_test = pd.DataFrame({
    "Country Name": ["Serbia", "Polonia", "España"],
    "Area (KM^2)": [88361.0, 312696.0, 505944.0],
    "Water (%)": [None, 2.60, 1.04],
    "Population (hab.)": [6690887.0, 41026068.0, 49570725.0],
    "Density (hab./km^2)": [115.00, 123.20, 97.47],
    "GDP ($)": [4.765400e+10, 8.421720e+11, 1.730000e+12],
    "Last Event Date": pd.to_datetime([
        "2006-06-03", "1989-12-29", "1978-12-29"
    ]),
    "Latitude (º)": [44.818, 52.230, 40.417],
    "Longitude (º)": [20.457, 21.011, -3.703]
})

wa = WikipediaAnalytics(["dummy"])
wa.df = df_test.copy()
test_code(wa)

# Prueba de la solución del alumno con ficheros
ARCHIVOS = [
    "serbia_es.html",
    "polonia_es.html",
    "espania_es.html",
]

wa = WikipediaAnalytics(ARCHIVOS)
wa.scrap()
test_code(wa)
```
