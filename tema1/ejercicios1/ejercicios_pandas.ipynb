# Ejercicios de Selección, Filtrado y Creación de Nuevos DataFrames

## Ejercicio 2

## Ejercicio 3
Crear un nuevo DataFrame con una columna adicional `fullname` que contenga el nombre completo (`firstname` + `lastname`).

```python
df['fullname'] = df['firstname'] + ' ' + df['lastname']
print(df)
```

## Ejercicio 4

## Ejercicio 5
Seleccionar y mostrar solo las filas donde el correo electrónico contenga `gmail.com`.

```python
gmail_users = df[df['email'].str.endswith('@gmail.com')]
print(gmail_users)
```

## Ejercicio 6
Crear un nuevo DataFrame con una columna adicional `age` que contenga la edad calculada a partir de la fecha de nacimiento.

```python
df_age = df.copy()

df_age['age'] = ((pd.Timestamp.now() - df_age.index).days / 365.25).astype(int)

# Se tienen en cuenta los años bisiestos con 365.25
# (365 + 365 + 365 + 366) / 4 = 365.25
print(df_age)
```

## Ejercicio 7
Seleccionar y mostrar solo las filas donde la edad sea mayor a 30 años.

```python
age_above_30 = df_age[df_age['age'] > 30]
print(age_above_30)
```

## Ejercicio 8

## Ejercicio 9
Crear un nuevo DataFrame con una columna adicional `domain` que contenga el dominio del correo electrónico.

```python
df_domain = df.copy()

df_domain['domain'] = df_domain['email'].str.split('@').str[1] # [0] es el usuario | [1] es el dominio
print(df_domain)
```

## Ejercicio 10
Seleccionar y mostrar solo las filas donde el dominio del correo electrónico sea `hotmail.com`.

```python
hotmail_domain = df_domain[df_domain['domain'] == 'hotmail.com']
print(hotmail_domain)
```

# Ejercicios de Agrupación, Ordenación y Unión
## Ejercicio 1
Agrupar por apellido y contar el número de ocurrencias de cada apellido.

```python
lastname_count = df.groupby('lastname').size()
'''
size() → cuenta todas las filas (incluyendo NaN)
count() → cuenta solo valores no nulos
value_counts() → ordena de mayor a menor automáticamente
'''
print(lastname_count)
```

## Ejercicio 2
Ordenar el DataFrame por fecha de nacimiento en orden ascendente.

```python
print(df.sort_index(ascending=True))
```

## Ejercicio 3
Unir dos DataFrames creados a partir del archivo CSV original.

```python
'''
concat() → apila DataFrames (vertical u horizontalmente)
merge() → une DataFrames por una columna clave común (como JOIN en SQL)
'''
```

```python
# Dos df a partir del original
df1 = df[df['lastname'] == 'Miller']
df2 = df[df['lastname'] == 'Smith']

# pd.concat = Unir verticalmente (concatear filas y columnas)
df_union = pd.concat([df1, df2])
print(df_union)
```

```python
# Dos df con cols diferentes
df1 = df[['firstname', 'lastname']]
df2 = df[['firstname', 'email']]

# Unir por columna común
df_merge = pd.merge(df1, df2, on='firstname')
print(df_merge)
```

## Ejercicio 6
Unir dos DataFrames utilizando la columna `id` como clave.
